# Resolução

```python
# os dados de iris estão em formato de array
# 1 transformar o array em um dataframe
import pandas as pd
```

```python
iris_df = pd.DataFrame(iris.data,columns=['sepal-length','sepal-width', 'petal-length', 'petal-width'])
```

```python
iris_df
```

```python
# 2 transformar o array com as classificações em um dataframe

class_df = pd.DataFrame(iris.target,columns=['class'])
```

```python
class_df
```

```python
# 3 atribuir o nome correto das classificações

class_df['class'] = class_df['class'].replace([0,1,2],['setosa','versicolor','virginia'])
```

```python
class_df
```

```python
# 4 juntar os dois dataframes

iris = pd.concat([iris_df,class_df],axis=1)
```
